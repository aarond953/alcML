```python
checker3='2nm_CD032_0.5_(.*)ul_(.*)ulethanol_(.*)ulwater_(.*)_(.*).txt'
df2=pd.DataFrame(index=None,columns=('alcohol_concentration','CD_amount','typ','file_name','spec'))
for key in dnormdict: 
  if re.search(checker3,key) is not None: 
    con=np.float(re.search(checker3,key).group(2))/(np.float(re.search(checker3,key).group(2))+np.float(re.search(checker3,key).group(3))) 
    am=np.float(re.search(checker3,key).group(1)) 
    typ=re.search(checker3,key).group(4)
    print(typ)
    dat=pd.DataFrame([[con,am,typ,key,dnormdict[key]]]) 
    dat.columns=df2.columns 
    df2=df2.append(dat,ignore_index=True) 
```
